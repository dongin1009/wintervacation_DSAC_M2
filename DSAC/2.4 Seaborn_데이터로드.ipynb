- Seaborn을 import하면 색상 등을 Matplotlib에서 제공하는 기본 스타일이 아닌 Seaborn에서 지정한 기본 스타일로 바꾼다.
- 따라서 동일한 Matplotlib 명령을 수행해도 Seaborn을 import 한 것과 하지 않은 플롯의 모양이 다르다.

# set 명령
- 색상, ticks 스타일 등 전반적인 plot 스타일을 Seaborn 기본 스타일로 바꾼다.
- sns.set(): empty, default값으로 초기화
---
# set_style 명령
- ticks 스타일만 바꿀 수 있다.
- darkgrid, whitegrid, dark, white, 그리고 ticks 스타일을 제공

```python
import numpy as np
data = np.random.randn(10)
sns.kdeplot(data)
```

```python
sns.set_style('ticks')
sns.kdeplot(data)
```

```python
sns.set_style('dark')
sns.kdeplot(data)
```

```python
sns.set_style('white')
sns.kdeplot(data)
```

```python
sns.set_style('darkgrid')
sns.kdeplot(data)
```

```python
sns.set_style('whitegrid')
sns.kdeplot(data)
```

```python
sns.set()
%matplotlib inline

sns.swarmplot(x="species", y="petal_length", data=iris)
```

# swarmplot
   - stripplot과 비슷하지만 데이터를 나타내는 점이 겹치지 않도록 옆으로 이동한다.

```python
# Swarm Plot
sns.swarmplot(x="species", y="petal_width", data=iris)
```

```python
sns.swarmplot(x="total_bill", y="tip", data=tips)
```

```python
sns.swarmplot(x="total_bill", y="sex", data=tips)
```

# stripplot
   - 마치 scatter plot처럼 모든 데이터를 점으로 그려준다.
   - jitter=True를 설정하면
       - 가로축상의 위치를 무작위로 바꾸어서 데이터의 수가 많을 경우에 겹치지 않도록 한다.

```python
# 카테고리 변수에 대한 구분
sns.stripplot(x="total_bill", y="sex", data=tips)
```

```python
sns.stripplot(x="total_bill", y="sex", data=tips, jitter=True)
```

# barplot
   - 카테고리 값에 따른 실수 값의 평균과 편차를 표시하는 기본적인 바 차트를 생성한다.
   - 평균은 막대의 높이
   - 편차는 에러바(error bar)로 표시

```python
# 카테고리 변수에 대한 구분
sns.barplot(x='day', y='total_bill', data=tips)
```
