```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
iris_data = sns.load_dataset("iris")
x = sns.PairGrid(iris_data)
x = x.map(plt.scatter)
```

```python
#Show a univariate distribution on the diagonal:
x = sns.PairGrid(iris_data)
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
# Color the points using a categorical variable:
x = sns.PairGrid(iris_data, hue = 'species')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
# Use a diffent style to show multiple histograms:
x = sns.PairGrid(iris_data, hue = 'species')
x = x.map_diag(plt.hist, histtype = 'step', linewidth = 3)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris_data, hue = 'species', vars = ['petal_length', 'petal_width'])
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
# Use diffent functions on the upper and lower triangles:

x = sns.PairGrid(iris_data)
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
x = x.add_legend()
```

```python
# Use diffent colors and markers for each categorical level:

g = sns.PairGrid(iris_data, hue="species", palette="Set2", hue_kws={"marker": ["o", "s", "D"]}) # o:o s:square D:diamond
g = g.map(plt.scatter, linewidths=1, edgecolor="w", s=40)
g = g.add_legend()
```
