하나의 그림에서 두가지 배경을 설정하기 위해서 axes_style()을 사용한다.
   - with 문 내에서 axes_style()을 사용하면,
   - plotparameter를 일시적으로 set할 수 있다.
   - with문을 빠져나오면서 일시적으로 set한 내용은 닫혀진다.
    

````python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
        
# 그림 1
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
    
# 그림 2
plt.subplot(212)
sinplot(-1)
````

- Scaling
     - plotting_context()
     - .set_context()
         * option:```paper```, ```notebook(디폴트)```, ```talk```, ```poster```
         * This affects things like the size of the labels, lines, and other elements of the plot,
         * but not the overall style.

````python
# 환경 재설정
sns.set()
````

````python
sns.set_style("whitegrid")
sns.set_context("paper")

sns.boxplot(x = "day", y="total_bill", data=tips)
````

````python
sns.set_context("poster")
sns.boxplot(x="day", y="total_bill", data=tips)
````

````python
sns.set_context("notebook")
sns.boxplot(x="day", y="total_bill", data=tips)
````

````python
sns.set_context("talk")
sns.boxplot(x="day", y="total_bill", data=tips)
````

# 그림의 크기가 달라지지 않는 것은 쥬피터에서 조정하기 때문이다.
- rc으로도 시각적 효과를 바꿀 수 있다.

````python
sns.set(style='whitegrid', rc={"grid.linewidth":1.5})
sns.set_context("poster", font_scale=2.5, rc={"lines.linewidth":5.0})
sinplot()
````

````python
sns.set(style='whitegrid', rc={"grid.linewidth":1.5})
sns.set_context("poster", font_scale=1.5, rc={"lines.linewidth":2.0})
sinplot()
````
